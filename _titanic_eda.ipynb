```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv', sep=',')
print(df.dtypes)
```

```python
print("women survival rate:",df[df['Sex']=='female']['Survived'].mean()*100)
print("Men survival rate:",df[df['Sex']=='male']['Survived'].mean()*100)
```

```python
print("Survival rate for class 1:",df[df['Pclass']==1]['Survived'].mean()*100)
print("Survival rate for class 2:",df[df['Pclass']==2]['Survived'].mean()*100)
print("Survival rate for class 3:",df[df['Pclass']==3]['Survived'].mean()*100)
```

```python
print(df.groupby('Pclass')['Survived'].mean()*100)
print(df.groupby('Sex')['Survived'].mean()*100)
```

```python
print(df.groupby('Pclass')['Fare'].mean())
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=['Cabin'])
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df.isnull().sum())
```

```python
print(df.nlargest(5,'Fare')[['Name','Fare','Pclass','Survived']])
```

```python
print(df.nsmallest(5,'Age')[['Name','Age','Pclass','Survived']])
```

```python
print(df.corr(numeric_only=True))
```

```python
print(df.groupby('Survived')[['Age', 'Fare']].mean())
```

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print(df[(df['Sex']=='male')&(df['Pclass']==3)]['Survived'].mean()*100)
```

```python
import matplotlib.pyplot as plt
survival_by_sex=df.groupby('Sex')['Survived'].mean()*100
survival_by_sex.plot(kind='bar',color=['steelblue','salmon'],edgecolor='black')
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate %')
plt.xticks(rotation=0)
plt.show()
```

```python
survival_by_class=df.groupby('Pclass')['Survived'].mean()*100
survival_by_class.plot(kind='bar',color=['orchid','gold','coral'],edgecolor='black')
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate %")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['Age'].plot(kind='hist',bins=20,color='steelblue',edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['Fare'].plot(kind='hist',bins=30,color=['salmon'],edgecolor='black')
plt.title("Fare Distribution of Passengers")
plt.xlabel("Fare")
plt.ylabel('Number of Passengers')
plt.show()
```

```python
avg_fare_per_port=df.groupby('Embarked')['Fare'].mean()
avg_fare_per_port.plot(kind='bar',color=['Gold','Coral','Orchid'],edgecolor='black')
plt.title("Average Fare Per Embarked Port")
plt.ylabel('Average Fare')
plt.xticks(rotation=0)
plt.show()
```
