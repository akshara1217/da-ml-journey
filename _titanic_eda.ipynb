# Titanic EDA — Exploratory Data Analysis

## Dataset
891 passengers, 12 columns. Source: Kaggle Titanic dataset.

## Key Findings
- Overall survival rate: 38.4%
- Women survived at 74% vs men at 18%
- 1st class survival: 63% vs 3rd class: 24%
- Higher fare strongly linked to survival
- Most passengers were aged 20-30
- Cherbourg port had highest average fare

## Techniques Used
Pandas filtering, groupby, missing data handling,
Matplotlib bar charts, histograms, scatter plots, pie chart

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print(df[(df['Sex']=='male')&(df['Pclass']==3)]['Survived'].mean()*100)
```

```python
import matplotlib.pyplot as plt
survival_by_sex=df.groupby('Sex')['Survived'].mean()*100
survival_by_sex.plot(kind='bar',color=['steelblue','salmon'],edgecolor='black')
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate %')
plt.xticks(rotation=0)
plt.show()
```

```python
survival_by_class=df.groupby('Pclass')['Survived'].mean()*100
survival_by_class.plot(kind='bar',color=['orchid','gold','coral'],edgecolor='black')
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate %")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['Age'].plot(kind='hist',bins=20,color='steelblue',edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['Fare'].plot(kind='hist',bins=30,color=['salmon'],edgecolor='black')
plt.title("Fare Distribution of Passengers")
plt.xlabel("Fare")
plt.ylabel('Number of Passengers')
plt.show()
```

```python
avg_fare_per_port=df.groupby('Embarked')['Fare'].mean()
avg_fare_per_port.plot(kind='bar',color=['Gold','Coral','Orchid'],edgecolor='black')
plt.title("Average Fare Per Embarked Port")
plt.ylabel('Average Fare')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
import matplotlib.pyplot as plt
average_age_by_class=df.groupby('Pclass')['Age'].mean()
average_age_by_class.plot(kind='bar',color=['gold','salmon','coral'],edgecolor='black')
plt.title("Average Age by Passenger Class")
plt.ylabel("Average Age")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Age'],df['Fare'],alpha=0.4,color='steelblue')
plt.title('Age v.s. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8,5))
died=df[df['Survived']==0]
survived=df[df['Survived']==1]
plt.scatter(died['Age'],died['Fare'],alpha=0.4,color='red',label='Died')
plt.scatter(survived['Age'],survived['Fare'],alpha=0.4,color='green',label='Survived')
plt.title("Age vs Fare - Coloured by Survival")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Survived'].value_counts().plot(kind='pie',
                                   labels=['Died','Survived'],
                                   colors=['salmon','steelblue'],
                                   autopct='%1.1f%%')
plt.title('Survival Distribution')
plt.ylabel(' ')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
plt.figure(figsize=(8,5))
class_1=df[df['Pclass']==1]
class_2=df[df['Pclass']==2]
class_3=df[df['Pclass']==3]
plt.scatter(class_1['Age'],class_1['Fare'],alpha=0.4, color='gold',label='Class 1')
plt.scatter(class_2['Age'],class_2['Fare'],alpha=0.4, color='blue',label='Class 2')
plt.scatter(class_3['Age'],class_3['Fare'],alpha=0.4, color='brown',label='Class 3')
plt.title('Age vs Fare based on Passenger Class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```
