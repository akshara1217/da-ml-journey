```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/PhilippeCodes/World-Happiness-Report-Data-Analysis/master/World%20Happiness%20Report.csv')

# Convert a Pandas column to NumPy array
happiness_scores = df['Happiness Score'].to_numpy()
print(type(happiness_scores))
print(happiness_scores[:5])

# NumPy operations
print("Mean:", np.mean(happiness_scores))
print("Std:", np.std(happiness_scores))
print("Median:", np.median(happiness_scores))
print("25th percentile:", np.percentile(happiness_scores, 25))
print("75th percentile:", np.percentile(happiness_scores, 75))
```

```python
q1= np.percentile(happiness_scores, 25)
q3= np.percentile(happiness_scores, 75)
iqr=q3-q1
lower=q1-(1.5*iqr)
upper=q3+(1.5*iqr)
outliers = happiness_scores[(happiness_scores > upper) | (happiness_scores < lower)]
print(outliers)
```
